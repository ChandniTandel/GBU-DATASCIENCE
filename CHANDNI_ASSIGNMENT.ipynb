```python
# Q.2- plot line diagram of active,confirmed,recovered,deceased cases district wise
```

```python
#LINE PLOT
```

```python
A = data.iloc[0:,1].values
C = data.iloc[0:,2].values
D = data.iloc[0:,3].values
R = data.iloc[0:,4].values
Z = data.iloc[0:,0]
plt.plot(Z, A, label="Active Cases", color= "BLUE")
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('Covid-19 cases')
plt.legend()
plt.show()
```

```python
plt.plot(Z, C, label="Confirmed cases",color= "PURPLE")
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('Covid-19 cases')
plt.legend()
plt.show()
```

```python
plt.plot(Z, D, label="Deceased cases",color= "GREEN")
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('Covid-19 cases')
plt.legend()
plt.show()
```

```python
plt.plot(Z, R, label="Recovered cases",color= "RED")
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('Covid-19 cases')
plt.legend()
plt.show()
```

```python
plt.plot(Z, A, label="active cases", color= "YELLOW")
plt.plot(Z, C, label="Confirmed cases",color= "RED")
plt.plot(Z, D, label="Deceased cases",color= "ORANGE")
plt.plot(Z, R, label="Recovered cases",color= "BLUE")
plt.xlabel('DISTRICT')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend() 
plt.show() 
```

```python
plt.bar(Z, D, label="deceased cases",color="PURPLE")
plt.xlabel('DISTRICT')
plt.ylabel('NO. OF CASES')
plt.title('COVID-19 cases')
plt.legend() 
plt.show()
```

```python
plt.bar(Z, R, label="Recovered cases",color="RED")
plt.xlabel('DISTRICT')
plt.ylabel('NO. OF CASES')
plt.title('COVID-19 cases')
plt.legend() 
plt.show()
```

```python
plt.bar(Z, A, label="Active cases", color= "red")
plt.bar(Z, C, label="confirmed cases",color="green")
plt.bar(Z, D, label="deceased cases",color="black")
plt.bar(Z, R, label="Recovered cases",color="orange")
plt.xlabel('DISTRICT')
plt.ylabel('NO. OF CASES')
plt.title('COVID-19 cases')
plt.legend() 
plt.show()
```

```python
#HISTOGRAM
```

```python
plt.hist(A, label=" ACTIVE CASES", color= "green")
plt.hist(C, label="CONFIRMED CASES",color= "black")
plt.hist(C, label="DECEASED CASES",color= "red")
plt.hist(C, label="RECOVERED CASES",color= "blue")
plt.xlabel('No. of cases')
plt.ylabel('Frequency')
plt.title('COVID-19 cases')
plt.legend() 
plt.show()
```

```python
#BOXPLOT
```

```python
covidcases = [A,C,D,R]
plt.boxplot(covidcases)
plt.title('covid-19 cases')
plt.xlabel('active confirmed deceased recovered ')
plt.ylabel('No. of cases')
plt.show()
```

```python
# Q.4- plot only active vs recovered cases for top 5 district having highest no.
```

```python
data.sort_values(['districtData/0/active', 'districtData/0/district'], ascending = False)
```

```python
sortcases = data.sort_values(['districtData/0/active', 'districtData/0/district'], ascending = False)
```

```python
sortcases.head(5)
```

```python
highestcases = sortcases.head(5)
```

```python
#BARGRAPH
```

```python
a = highestcases.loc[:, "districtData/0/active"]
r = highestcases.loc[:,"districtData/0/recovered"]
plt.bar(a,r, width = 450, color="red")
plt.xlabel("Active cases")
plt.ylabel("Recovered cases")
plt.title("covid-19 cases in top 5 districts")
plt.show()
```

```python
#SCATTER PLOT
```

```python
plt.scatter(a, r)
plt.xlabel("Active cases")
plt.ylabel("Recovered cases") 
plt.title("Scatter Plot of active and recovered cases of top 5 districts") 
plt.tight_layout() 
plt.show() 
```

```python
#BOX PLOT
```

```python
Covidcases = [a, r]
plt.boxplot(Covidcases)
plt.title('covid-19 cases of top 5 districts')
plt.xlabel(' active recovered ')
plt.ylabel('No. of cases')
plt.show()
```

```python
#LINE PLOT
```

```python
plt.plot(a, r, color= "blue")
plt.xlabel('Active cases')
plt.ylabel('Recovered cases')
plt.title('COVID-19 cases in top 5 districts') 
plt.show()
```

```python
#HISTOGRAM
```

```python
plt.hist(a, label= "Active cases", color = "blue")
plt.hist(r, label= "recovered cases", color = "red")
plt.title('COVID-19 cases in top 5 districts') 
plt.xlabel("No. of covid cases")
plt.ylabel(" frequency")
plt.show()
```
